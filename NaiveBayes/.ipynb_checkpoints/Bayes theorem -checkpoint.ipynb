## Bayes Theorem:


                    P(A/B) = P(B/A) * P(A)
                             --------------
                                 P(B)

### Let us understand the theorem with example.

Suppose we have two machines that produces screw drivers-
M1 = 30 screw drivers
M2 = 20 screw drivers

Out of all the produced 1% are defective.
We can see out of all defective parts 50% came from machine 1 and other 50% from machine two.

Question - 
What is the probability that a path produced by machine 2 is defective.
 
 Total screwdrivers produced = 30+20 = 50
 So P(M1) = 30/50
    P(M2) = 20/50
 
 Also P(Defect) = 1%
 
 Defected-
 Probabilty of M1 given it is defected-
 
     P(M1 | Defect) = 50%
     
 Probability of M2 given it is defected-
 
     p(M2 | Defect) = 50%
     
Our question is to find the probabilty that a part produced by mach2 is defective-

    P(Defect | M2) 
    
Now using bayes therorem-
    
    P(Defect/Mach2) = P(Mach2 | Defect) * P(Defect)
                      ------------------------------
                              P(Mach2)
                              
                              
                    = 0.5 * 0.01
                     ------------  = 0.0.125
                         0.4
