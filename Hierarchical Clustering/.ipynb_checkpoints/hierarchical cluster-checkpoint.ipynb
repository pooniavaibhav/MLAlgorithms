## There are two types of HC-
#### Agglomerative - Bottom up
#### Divisive - from up to down

### Agglomerative-
##### Step -1 -Make each data point a single -point cluster that forms N cluster.
##### Step -2 -Take the two closest data points and make them one cluster that forms N-1 clusters.
##### Step -3 -Take the two closest clusters and make them one cluster that forms N-2 clusters.
##### Step -4 -Repeat step 3 untill there is only one cluster.
##### FINISH

### The way HC works is it maintains in memory the process to reach the final step. That memory is called the Dendrogram.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Mall_Customers.csv")
```

```python
X = data.iloc[:,[3,4]].values
```

### Using dendograms to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X,method = 'ward'))
```

### Fitting hierarchical clustering to the mall dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5,affinity='euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```
