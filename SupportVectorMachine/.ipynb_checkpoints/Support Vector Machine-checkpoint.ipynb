### SVM

Suppose we have multiple points poltted on the graph.It is a classification problem so  these groups can be devided by line.So SVM use to draw a line between these classes of points and at the same time it has the maximum margin.

```python
from IPython.display import Image
Image(filename = 'graph.png')
```

#### The two points that helps in maintaining the maximum margin are called Support Vectors. 
We are calling them vectors as when there will be more variables this problem will become multi dimensional so calling them vectors as it will have magnitude and direction as well.

#### Maximum Margin Hyperplane:
The centre line or plane is called Maximum Margin Hyperplane or Maximum Margin Classifier.  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
dataset.head()
```

```python
x = dataset.iloc[:,[2,3]].values
y = dataset.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train) 
x_test = sc.fit_transform(x_test)
```

#### Fitting classification model

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
classifier.fit(x_train,y_train)
```

```python
classifier.predict(x_test)
```
