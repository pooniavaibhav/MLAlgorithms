### Polynomial Regression:
The eqaution for polynomial linear equation is:

        y = b0 + b1x1 + b2x1^2 + ....+ bnx1^n
        
                here x1^2 = square of x
                we have a same varibale xn but with different power.
                
  As shown below when you plot this equation instead of straight line you get a curve this  is because this power on independent var gives it that parabolic effect. 
  
  In use cases line pendemic spread across population.

Question - Since it has powers in your equation why is it still called linear when it has polynomial with it.

Answer - When we talk about linear -non linear we talk about the coefficiants. And these coefficients are still linear bcz what we need is b0,b1...bn they do not have any power.

So polynomial linear regression is called the special case of multiple linear regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("Position_Salaries.csv")
```

```python
data_set.head()
```

```python
X = data_set.iloc[:,-2].values
Y = data_set.iloc[:,-1].values
```

### Always remember take X as a matrix and Y as a vector:

```python
X = data_set.iloc[:,1:2].values
print(X)
```

### Split dataset into training and test
since data set is very small we dont need to split our data.
Also we do not require to feature scale our data bcz polynomial regression expects polynomial term.  

### So let us create both linear and polynomial regression model to compare result: 

#### Linear model

```python
from sklearn.linear_model import LinearRegression
regressor_linear  = LinearRegression()
regressor_linear =  regressor_linear.fit(X,Y)
```

#### Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(X)
regressor_linear2 =LinearRegression()
regressor_linear2 = regressor_linear2.fit(x_poly,Y)
```

#### Visualising the Linear Regression results

```python
plt.scatter(X,Y,color = 'red')
plt.plot(X,regressor_linear.predict(X),color = 'blue')
plt.title('Truth or Bluff(LinearRegression)')
plt.xlabel('level')
plt.ylabel('Salary')
plt.show()
```

#### Visualize polynomial regression result:

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X,Y,color = 'red')
plt.plot(X_grid,regressor_linear2.predict(poly_reg.fit_transform(X_grid)),color = 'blue')
plt.title('Truth or Bluff(PolynomialRegression)')
plt.xlabel('level')
plt.ylabel('Salary')
plt.show()
```
