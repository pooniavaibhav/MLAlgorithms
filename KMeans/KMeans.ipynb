### Kmeans
Suppose we have some sample data.We dont have our dependent variable.So can we identify
some groups among our variables.

##### Steps:
Step1 - Choose the number K of clusters.
Step2 - Select at random K points , the centroids (not necessarily from your data set.)
Step3 - Assign each data point to the closest centroid that forms K Cluster. 
Step4 - Compute and place the new centroid of each cluster.
Step5 - Reassign each data point to the new closest centroid.If any reassignment took place go to step 4,otherwise go to FIN.

### Elbow Method:
In graph look for the the elbow that is forming or the point after which the difference in values is very less . 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/home/vaibhav/AIGit/KMeans/Mall_Customers.csv")
```

```python
x= dataset.iloc[:,3:5].values
```

### Optimal number of clusters using elbow method:

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1, 11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter = 300,n_init=10)
    kmeans = kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title("Elbow method")
plt.xlabel("Number of cluster")
plt.ylabel("WCSS")
plt.show()
```

### Applying kmeans:

```python
kmeans=KMeans(n_clusters=5,init='k-means++',max_iter = 300,n_init=10)
y_kmeans = kmeans.fit_predict(x)
```
