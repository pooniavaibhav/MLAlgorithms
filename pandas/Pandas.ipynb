### Grouping and Aggregating:

```python
data["ConvertedComp"].median()
#if we directly type median on the dataframe it will give median of every available numerical value.
data.median()
#we can get a bunch of insight in data uding describe keyword.
data.describe()
#count give us non missing rows REMEMBER non NA rows.
data["ConvertedComp"].count()
#Suppose we want value wise count of data we use value count-
data["Hobbyist"].value_counts()
```

```python
#we can get the percentage of value counts using normalize = True.
data["SocialMedia"].value_counts(normalize = True)
```

```python
#suppose we want to group above data according to country.
#Group by involves a combination of three-
#-Splitting the object
#-Applying a function.
#-Combine the result.

filt = data['Country'] == 'India'
data.loc[filt]['SocialMedia'].value_counts()

#this can be done using group by but country wise-
country_grp = data.groupby(['Country']) 
country_grp['SocialMedia'].value_counts()
#you can even get country wise data.
country_grp['SocialMedia'].value_counts().loc['India']
```

```python
#we can use multiple aggregation functions on our data-
country_grp = data.groupby(['Country']) 
country_grp['ConvertedComp'].agg(['median','mean'])
```

```python
#.sum() works with numerical as well as boolean data. it will count False as 0 and 1 for True.
#we csn not use str() methods with aggregation so we need to use apply()
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
df3 ={'first':['suniel','salman','amir','shahrukh','Missing'],'last':['shetty','khan','khan', None,'Missing'],'email':['abc@gmail.com','def@gmail.com','NA', None,'Missing']}
df3 = pd.DataFrame(df3)
```

```python
#dropna() paramters
#It has two parameters-
#-axis :it tells if you want to drop row or a column e.g 'index' for rows and 'column' for columns. 
#-how  :it tells in which way you want to remoce the data e.g how = 'all' will only delete 
#if all the values of that record is missing.
df3.dropna(axis = 'index',how  = 'all') # no changes as we dnt have any record with all values missing.
```

```python
#suppose in my all data i wanted to remove a record only if email adress in empty:
#this can be achieved by using subset-
df3.dropna(axis='index', how='all', subset=['email'])
```
